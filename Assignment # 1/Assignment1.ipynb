```python
# 04 Write a python program that accepts the radius of a circle from the user and compute the area.

r = input(f'Type the radius of the circle here! {int()}')
pi = 3.142
area_of_the_circle = float(pi) * int(r) ** 2
print(f'Area of the circle is {area_of_the_circle}')
```

```python
# 05 Write a python program which accepts the user's first and last name and print them in reverse
# order with space between them.

your_first_name = input(f'Type your first name here!{str()}')
your_last_name = input(f'Type your last name here!{str()}')

first = your_first_name
second = your_last_name
full_name = f'{second} {first}'
print(full_name)
```
